```python
import pandas as pd
import numpy as np
```

```python
ls=pd.read_csv('../inside_airbnb_data/listings.csv')
```

## Listings Clean

```python
import re

ls.drop(columns= ['id', 'host_id','host_name', 'neighbourhood_group', 'last_review','license','reviews_per_month'], inplace=True)

def extract_info(text):
    # Extract rating
    rating_match = re.search(r'★(\d+\.\d+)', text)
    rating = float(rating_match.group(1)) if rating_match else 'unknown'
    
    # Extract number of bedrooms
    bedroom_match = re.search(r'(\d+)\sbedrooms?', text)
    bedrooms = int(bedroom_match.group(1)) if bedroom_match else 'unknown'
    
    # Extract number of baths
    bath_match = re.search(r'(\d+)\s(private\s)?baths?', text)
    baths = int(bath_match.group(1)) if bath_match else 'unknown'
    
    return rating, bedrooms, baths
```

```python
ls[['rating', 'bedrooms', 'baths']] = ls['name'].apply(lambda x: pd.Series(extract_info(x)))
ls.drop(columns='name', inplace=True)
ls
```

```python
ls.to_csv('listings_cleaned.csv', index=False)
```
